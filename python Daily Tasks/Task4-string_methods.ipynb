**3Q.What is the difference between string.find() and string.index() ? Explain with examples.**


```python
string_input="rain is falling"
print("result of string_input.find():",string_input.find("i"))
print("result of string_input.index():",string_input.index("i"))
#Both gives same result if the given character is found in the input string

print("result of string_input.find():",string_input.find("z"))#if the given character is not found in input then it will give result as -1
print("result of string_input.index():",string_input.index("z"))#if the given character is not found in input then it will give result as ERROR

```
