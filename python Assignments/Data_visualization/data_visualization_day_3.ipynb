###### 1. What is Histogram chart ?

#### write here

**1. Histogram**
- 
What it shows: A histogram represents the frequency distribution of a single numerical variable, breaking the data into bins and showing how many  -    data points fall within each bin.nsformation).

**What we understand:**

- Skewness (left or right skewed distribution).
- Concentration of data points (high or low frequency bins).
- Spread and range of the variable.
- Presence of outliers (extreme values).
  
**How it helps in analysis:**

- Helps you understand the shape and spread of the data.
- Allows you to detect if data is normally distributed or not.
- Can highlight any potential need for transformation (e.g., log transformation).

##### 2. What is box plot ?

### write here
**What it shows:**
- Boxplots summarize the distribution of a variable and highlight potential outliers using quartiles and whiskers.

**What we understand:**

- Median (central tendency).
- IQR (Interquartile Range): The range between Q1 (25th percentile) and Q3 (75th percentile).
- Outliers: Shown as points outside the whiskers.

**How it helps in analysis:**

- Detects outliers and extreme values.
- Visualizes the spread and symmetry of data.
- Useful for comparing distributions across multiple groups.

##### 3. What are the plots we can use for univeriate analysis. 

### write here
- **Histogram**: Shows the distribution of continuous numerical data.
- **Box Plot**: Provides a summary of a variable's distribution, including outliers.
- **Violin Plot**: Combines the features of a box plot and a KDE, showing both the distribution and the density of the data.
- **Density Plot (KDE)**: Shows the continuous probability density function of a variable.
- **Bar Plot**: Used for categorical data to show the frequency of each category.
- **Pie Chart**: Displays the relative proportions of categories in a dataset.
- **Count Plot**: Displays the count of occurrences of each category, specifically for categorical variables.
- **Strip Plot:** Shows the distribution of categorical variables by plotting individual data points.

##### 4. What are the plots we can use for bivariate analysis.

### code here
- **Scatter Plot**: Shows relationships between two continuous variables.
- **Line Plot**: Shows the relationship between two continuous variables over a time-based or ordered domain.
- **Heatmap (Correlation Matrix)**: Visualizes the correlation between multiple continuous variables using color gradients.
- **Pair Plot (Scatterplot Matrix):** Shows relationships between all pairs of variables in a dataset.
- **Bar Plot (Categorical vs Continuous):** Compares the means of a continuous variable across different categories.
- **Box Plot (Categorical vs Continuous):** Shows the distribution of a continuous variable across different categories.
- **Strip Plot (Categorical vs Continuous):** Displays individual data points to visualize the distribution across categories.


###### 5. What are plot we can use for multivariate analysis.
