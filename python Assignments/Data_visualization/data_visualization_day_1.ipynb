#### 1. create a bar plot for count the frequency of the three species.

```python
### code here
df['Species'].value_counts().plot(kind="bar")
```

##### 2. Create a pie chart to show the percentage of species.

```python
### code here
df['Species'].value_counts().plot(kind="pie",autopct="%.2f%%")
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=df, kind='scatter')
plt.show()
```

##### 4. Create a scatter plot and take hue = "Species"

```python
### code here
sns.scatterplot(x=df['SepalLengthCm'],y=df['PetalLengthCm'],hue=df['Species'],data=df)
```

### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
### code here
#univariate
df['SepalWidthCm'].plot(kind="box")
```

```python
#bivariate Analysis
df[['SepalWidthCm','PetalWidthCm']].plot(kind='box')
```

###### 6. Create a Strip plot 

```python
### code here
sns.stripplot(y='SepalWidthCm',x='Species',data=df)
```

###### 7. Create a Violin Plot.

```python
### code here
sns.violinplot(x='SepalWidthCm',y='Species',data=df)
```

##### 8. Create a pair chart.

```python
### code here
sns.pairplot(data=df,hue='Species')
```

##### 9. Create a Heat Map

```python
### code here
df1=df.select_dtypes(include="number")
df1=df1.corr()
sns.heatmap(df1)
plt.show()
```

### 10. Create a distribution plot for all numerical colomns.

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(5,5))
axes[0][0].set_title("sepalLengthCm")
axes[0][1].set_title("SepalWidthCm")
axes[1][0].set_title('PetalLengthCm')
axes[1][1].set_title("PetalWidthCm")
df["SepalLengthCm"].plot(kind="density",ax=axes[0][0])
df["SepalWidthCm"].plot(kind="density",ax=axes[0][1])
df["PetalLengthCm"].plot(kind="density",ax=axes[1][0])
df["PetalWidthCm"].plot(kind="density",ax=axes[1][1])
```
