###### 17. Write a Pandas program to get the current date, oldest date and number of days between Current date and oldest date of Ufo dataset.

```python
#### code here
dataset['Date_time'] = pd.to_datetime(dataset['Date_time'],format='%d/%m/%Y %H:%M', errors='coerce')
```

```python
oldest_date=dataset['Date_time'].min()

# Calculate the number of days between the current date and the oldest date

days_difference = (today_date - oldest_date).days

# Output the results
print("Current Date:", today_date.date())
print("Oldest Date:", oldest_date.date())
print("Number of Days Between Current Date and Oldest Date:", days_difference)
```
